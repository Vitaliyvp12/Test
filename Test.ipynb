```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Завантаження даних з файлу CSV
df = pd.read_csv('C:/Users/VV/OneDrive/Документы/product_sales_test.csv')
```

```python
# Додавання стовпця "Сума продажу" з обчисленням
df['Amount of sale'] = df['Units Sold'] * df['Unit Price']

# Обчислення загальної суми продажів для кожного продукту
total_sales = df.groupby('Product Name')['Amount of sale'].sum()
```

```python
# Побудова стовпчикової діаграми
plt.figure(figsize=(10, 6))
total_sales.plot(kind='bar', color='skyblue')
plt.title('Total sales for each product')
plt.xlabel('Product Name')
plt.ylabel('Amount of sale')
plt.xticks(rotation=45)
plt.show()
```
